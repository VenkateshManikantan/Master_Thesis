```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches
```

```python
data_large = pd.read_csv("data/new_all_features.csv")
data_small = pd.read_table("data/sequence_ogt_topt_features2_JF_19Jun22.tab")
```

```python
data_large
```

```python
target_0 = data_large["TEMPERATURE OPTIMUM"]
target_1 = data_small["seq_TEMP"]
```

```python
target10 = target_0.append(target_1)
```

```python
target10
```

```python
sns.distplot(target_0, hist=True, rug=True,color = "red")
sns.distplot(target_1, hist=True, rug=True,color = "blue")
sns.distplot(target10, hist=True, rug=True,color = "yellow")
plt.title('Density Plot of T-OPT from ENZYME Data 1 and ENZYME Data 2')
plt.xlabel('T-OPT')
blue_patch = mpatches.Patch(color='blue', label='Enzyme Data 1')
red_patch = mpatches.Patch(color='red', label='Enzyme Data 2')
green_patch = mpatches.Patch(color='yellow', label='Enzyme Data 1+2')
plt.legend(handles=[blue_patch,red_patch,green_patch])
plt.show()
```

```python
sns.distplot(target10, hist=True, rug=True,color = "red")
plt.title('Density Plot of T-OPT from ENZYME Data 1 + ENZYME Data 2')
plt.xlabel('T-OPT')
red_patch = mpatches.Patch(color='red', label='Enzyme Data 1+2')
plt.legend(handles=[red_patch])
plt.show()
```

```python
sns.swarmplot(target_0,color = "red")
sns.swarmplot(target_1,color = "blue")
plt.title('Density Plot of T-OPT from ENZYME Data 1 and ENZYME Data 2')
plt.xlabel('T-OPT')
blue_patch = mpatches.Patch(color='blue', label='Enzyme Data 1')
red_patch = mpatches.Patch(color='red', label='Enzyme Data 2')
plt.legend(handles=[blue_patch,red_patch])
plt.show()
```
